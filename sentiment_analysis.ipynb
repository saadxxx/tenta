```python
from transformers import pipeline

# 加载情感分析管道
classifier = pipeline("sentiment-analysis")
```

```python
# 输入文本
text = "I love using Hugging Face models, they are amazing!"

# 预测情感
result = classifier(text)
print(result)
```

```python
texts = [
    "I hate Mondays, they are the worst!",
    "This is the best day of my life!",
    "The weather is okay, not too bad."
]

results = classifier(texts)
for text, result in zip(texts, results):
    print(f"Text: {text}")
    print(f"Sentiment: {result['label']}, Confidence: {result['score']:.4f}")
    print()
```

```python
from datasets import load_dataset

# 加载情感分析数据集（例如IMDB电影评论）
dataset = load_dataset("imdb")

# 查看数据集内容
print(dataset["train"][0])  # 第一条数据
```

```python
sample_text = dataset["train"][0]["text"]
result = classifier(sample_text)
print(f"Text: {sample_text}")
print(f"Sentiment: {result[0]['label']}, Confidence: {result[0]['score']:.4f}")
```
